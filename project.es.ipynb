#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
precio_maximo = ds["price"].max()

casa_mas_cara = ds[ds["price"] == precio_maximo].iloc[0]

direccion = casa_mas_cara["realEstate_name"]
precio = casa_mas_cara["price"]

print(f"La casa con dirección en {direccion} es la más cara y su precio es de {precio} USD")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
ds["price"] = ds["price"].astype(float)

precio_minimo = ds["price"].min()

casa_mas_barata = ds[ds["price"] == precio_minimo].iloc[0]

direccion = casa_mas_barata["address"]  
precio = casa_mas_barata["price"]

print(f"La casa con direccion en {direccion} es la mas barata y su precio es {int(precio)} USD")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# TODO
ds["surface"] = ds["surface"].astype(float)
ds_limpio = ds[ds["surface"] > 0]

superficie_maxima = ds_limpio["surface"].max()
casa_mas_grande = ds_limpio[ds_limpio["surface"] == superficie_maxima].iloc[0]

direccion_grande = casa_mas_grande["address"]
superficie_grande = casa_mas_grande["surface"]

superficie_minima = ds_limpio["surface"].min()
casa_mas_pequena = ds_limpio[ds_limpio["surface"] == superficie_minima].iloc[0]

direccion_pequena = casa_mas_pequena["address"]
superficie_pequena = casa_mas_pequena["surface"]

print(f"La casa más grande está ubicada en {direccion_grande} y su superficie es de {int(superficie_grande)} metros")
print(f"La casa más pequeña está ubicada en {direccion_pequena} y su superficie es de {int(superficie_pequena)} metros")
```

#### Ejercicio 06. Elimina los valores nulos (NAs) del dataset, si aplica (★★☆)

Despues de eliminar los valores nulos, compara el tamaño del DataFrame antes y después de la eliminación.


```python
# TODO
ds_limpio = ds.copy()

ds_limpio["price"] = ds_limpio["price"].fillna(ds_limpio["price"].median())
ds_limpio["surface"] = ds_limpio["surface"].fillna(ds_limpio["surface"].median())
ds_limpio["rooms"] = ds_limpio["rooms"].fillna(ds_limpio["rooms"].median())
ds_limpio["bathrooms"] = ds_limpio["bathrooms"].fillna(ds_limpio["bathrooms"].median())

ds_limpio["realEstate_name"] = ds_limpio["realEstate_name"].fillna("Desconocido")
ds_limpio["phone_realEstate"] = ds_limpio["phone_realEstate"].fillna("Sin telefono")

print(ds.shape)
print(ds_limpio.shape)
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

filtro = ds["level5"] == "Arroyomolinos (Madrid)"
precios = ds[filtro]["price"]

plt.hist(precios, bins=20)
plt.title("Distribución de precios en Arroyomolinos (Madrid)")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")

plt.show()
```

La mayoría de los precios se concentran entre aproximadamente 180,000 y 350,000, lo que indica que la mayor parte de las viviendas en Arroyomolinos (Madrid) tienen un precio medio dentro de ese rango.

También se observan algunos valores más altos que llegan hasta cerca de 600,000, lo que sugiere la presencia de propiedades más caras (posibles outliers).

Además, aparece un valor muy bajo cercano a 0, lo cual podría ser un error en los datos o un valor atípico.

En general, la distribución está ligeramente sesgada hacia la derecha, algo común en datos de precios inmobiliarios.
